```python
#skiprow -> ignora as primeiras linhas
#skipfooter -> ignora as últimas linhas

df = pd.read_excel('C:/Users/mbrito/Desktop/Data Science/Curso LabHacker/CINEMA NO BRASIL.xlsx', skiprows=2, skipfooter=16)
```

```python
#Exibe as COLUNAS DO DATAFRAME

df.columns.str.upper()
```

```python
#RENOMEIA AS COLUNAS 

df.columns= ['ANO DE EXIBIÇÃO', 'NOME DO FILME', 'CPB/ROE', 'GÊNERO', 'PAÍS(ES) PRODUTOR(ES)', 'NACIONALIDADE', 'LANÇAMENTO', 'DISTRIBUIDORA', 'ORIGEM DISTRIBUIDORA', 'PÚBLICO', 'RENDA (R$)']
```

```python
df.head()
```

```python
# BUSCAR VALOR ESPECÍCO NA COLUNA
# sinal == é para comparação e não atribuição

df[df['LANÇAMENTO'] == 'Relançamento'].head()
```
